**Etapa 2 – Processo de Normalização da Matriz de Decisão**

> **Padronizando a Matriz de Decisão**



```python
# Ajustando os valores da coluna "MIN"
from pandas.core.internals.construction import ma
auxLinhaCriterios = matrizDecisao[matrizDecisao['Alternativas'] == 'Tipo:']
for (columnName, columnData) in auxLinhaCriterios.iteritems():
  columnData = columnData.values
  if columnData == 'MIN':
    auxData = matrizDecisao[matrizDecisao['Alternativas'] != 'Tipo:']
    auxData1 = auxData[columnName].values
    auxData1 = 1/auxData1
    auxData[columnName] = auxData1
    matrizDecisao = auxData
display(matrizDecisao)
```



> **Normalizando a Matriz de Decisão**



```python
# Apagando a coluna que contém as alternativas
auxColumns = matrizDecisao.drop(columns = ['Alternativas'])
# Gerando variável do tipo lista para guardar as informações
auxData = []
for coluna in auxColumns.columns:
  # Obtendo a soma de cada coluna da matriz
  auxSoma = matrizDecisao[coluna].sum()
  # Dividindo o valor de cada célula pela soma de sua coluna
  auxColuna = matrizDecisao[coluna]/auxSoma
  ## Obtendo os valores contidos nas colunas
  auxColuna = auxColuna.values
  ## Incluindo cada valor calculado e obtido na lista
  auxData.append(auxColuna)
  ## Convertendo a lista em DataFrame e transpondo - o
  matrizNormalizada = pd.DataFrame(auxData)
display(matrizNormalizada.T)
```

**Etapa 3 – Calculando a média das alternativas em cada critério**



```python
# Recebendo a matriz de decisão normalizada e transportada
matrizNormalizadaT = matrizNormalizada.T

# Calculando a média dos critérios
auxMediaLinhas = []
for i in matrizNormalizadaT.columns:
  auxMedia = matrizNormalizadaT[i].mean()
  auxMediaLinhas.append(auxMedia)
  mediaMatriz = pd.DataFrame(auxMediaLinhas)
  mediaMatriz.columns = ['Media']
display(mediaMatriz)
```

**Etapa 4 – Cálculo do desvio padrão dos critérios com base na amostra das alternativas**

```python
# Calculando o Desvio Padrão dos critérios
auxStdLinhas = []
for i in matrizNormalizadaT:
  auxStd = matrizNormalizadaT[i].std()
  auxStdLinhas.append(auxStd)
  stdMatriz = pd.DataFrame(auxStdLinhas)
  stdMatriz.columns = ['Desvio_Padrao']
display(stdMatriz)
```



> **Gerando Fator Gaussiano**


```python
# Gerando dataframe com a média e Desvio Padrão dos critérios
matrizMediaStd = mediaMatriz.join(stdMatriz)
matrizMediaStd = pd.DataFrame(matrizMediaStd)
display(matrizMediaStd)
```

```python
# Obtendo o fator Gaussiano
matrizMediaStdT = matrizMediaStd
auxGaussLinhas = []
for linha in matrizMediaStdT:
  auxLinha = matrizMediaStdT['Desvio_Padrao']/matrizMediaStdT['Media']
auxGaussLinhas.append(auxLinha)
fatorGaussiano = pd.DataFrame(auxGaussLinhas).T
fatorGaussiano.columns = ['Fator Gaussiano']
display(fatorGaussiano)
```

```python
somaFatorGaussiano = fatorGaussiano.sum()
display(somaFatorGaussiano)
```

```python
# Normalizando o fator Gaussiano
auxFatorNorm = []
for i in fatorGaussiano:
  auxFator = fatorGaussiano[i]/somaFatorGaussiano[0]
  auxFatorNorm.append(auxFator)
  fatorGaussianoNormalizado = pd.DataFrame(auxFatorNorm)
  fatorGaussianoNormalizadoC = fatorGaussianoNormalizado
#fatorGaussianoNormalizadoC.columns = matrizDecisaoCriterios = matrizDecisaoCriterios.delete(0)
display(fatorGaussianoNormalizadoC)
```

**Etapa 5 – Ponderando a matriz de decisão**

```python
# Transportando a matriz
matrizNormalizada = matrizNormalizada.T

# Gerando variável do tipo lista para guardar as informações abaixo
aux_result = []
matrizNorm = pd.DataFrame()
for i in matrizNormalizada:
  auxFator = fatorGaussianoNormalizado[i].unique()
  auxFator = np.float64(auxFator)
  matrizNorm[i] = matrizNormalizada[i].mul(auxFator)
display(matrizNorm)
```

```python
auxRankLinhas = []
matrizNorm = matrizNorm.T
for i in matrizNorm:
  auxRank = matrizNorm[i].sum()
  auxRankLinhas.append(auxRank)
  rankMatriz = pd.DataFrame(auxRankLinhas)
  rankMatriz.columns = ['Rank']
auxMatrizDecisao = pd.DataFrame(alternativas)
rankMatriz.sort_values(by = ['Rank'], ascending=False)
display(rankMatriz)
```
